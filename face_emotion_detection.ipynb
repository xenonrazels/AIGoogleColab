#**It's Boring right with these numerical value Let's massage our result with HAR_CASCADE Algorithm and Beautifully equipped Open-cv tools**
###(Note :viola jonas found this har_cascade algorithm:- which works for finding edges and gives output || Underhood HAR_CASCADE ML algorithms is used)

```python
#time for recognizing the face (face detection )
faceRegion=cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_frontalface_default.xml')
gray=cv2.cvtColor(perfectimg,cv2.COLOR_BGR2GRAY)
faces=faceRegion.detectMultiScale(gray,1.1,4)

# Time for drawing rectangle
#using loop if there are many faces in a pic
for (x,y,w,h) in faces:
  cv2.rectangle(perfectimg,(x,y),(x+w,y+h),(0,245,0),2)
```

#**Lets write emotion text display also, then we will show both in result.** 

```python
font=cv2.FONT_HERSHEY_COMPLEX
cv2.putText(perfectimg,
            predictPerfectImg['dominant_emotion'],
            (100,100),
            font,3,
            (0,0,255),
            2,
            cv2.LINE_AA,False
            )
```

#**CONCLUSION  : DeepFace works but it's result is not sastisfactionary.In next, We are going to train  our deep learnig model then we will check with same datasets. Then, we will compare our model with deepface model**

```python

# Moving towards the real time data emotion
!pip install deepface
from deepface import DeepFace
import cv2
faceReg=cv2.CascadeClassifier(cv2.data.haarcascades +'haarcascade_frontalface_default.xml')
cap=cv2.VideoCapture(1)
if not cap.isOpened():
  cap=cv2.VideoCapture(0)
# if not cap.isOpened():
  # raise IOError("Cannot open webcam")
while True:
  ret, frame = cap.read()
  result = DeepFace.analyze(frame, actions = ['emotion'])
  gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
  faces=faceReg.detectMultiScale(gray,1.1,4)
  for (x,y,w,h) in faces:
    cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),2)
    font=cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(frame,result['dominant_emotion'],
                (50,50),
                font,3,
                (0,0,255),
                2,cv2.LINE_4)
    cv2.imshow('Original Video',frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
      break

cap.release()
cv2.destroyAllWindows()
```
